# Pipeline ETL construído

A principal razão para estabelecer um pipeline ETL (Extrair, Transformar, Carregar) é aumentar a eficiência e a precisão do processamento de dados, o que é particularmente importante para empresas que lidam regularmente com grandes volumes de dados. O estabelecimento deste pipeline ETL pode automatizar todo o procedimento de processamento de dados, economizando assim tempo e recursos valiosos.

```python
# Based on the provided details, I'll update the etl_pipeline function to include the additional transformations
def etl_pipeline(file_path):
    import pandas as pd
    import numpy as np

    # Extract phase: Load the data from the given file path
    df = pd.read_csv(file_path)

    # Transform phase: Perform data cleaning and preprocessing
    # Drop rows where 'CustomerID' is NaN
    df = df.dropna(subset=['CustomerID'])

    # Fill NaN values in 'Discount_pct' with 'No Discount'
    df['Discount_pct'] = df['Discount_pct'].fillna('No Discount')

    # Drop unnecessary columns
    columns_to_delete = ["Unnamed: 0", "Transaction_Date", "Product_Description", "Coupon_Code"]
    df = df.drop(columns_to_delete, axis=1)

    # Rename columns
    df.rename(columns={
        "Tenure_Months": "Usage_Duration_Months",
        "Product_SKU": "Inventory_Number",
        "GST": "Tax",
        "Discount_pct": "Discount_%"
    }, inplace=True)

    # Data type conversions
    df['CustomerID'] = df['CustomerID'].astype('int').astype('string')
    df['Transaction_ID'] = df['Transaction_ID'].astype('int').astype('string')
    df['Tax'] = df['Tax'].astype('string')
    df['Month'] = df['Month'].astype('string')
    df['Quantity'] = df['Quantity'].astype('int')
    df['Date'] = pd.to_datetime(df['Date'])
    df['Gender'] = df['Gender'].astype('category')
    df['Coupon_Status'] = df['Coupon_Status'].astype('category')
    df['Discount_%'] = df['Discount_%'].astype('category')

    # Calculate sub_sum
    df['sub_sum'] = df['Quantity'] * df['Avg_Price']

    # Create Spend_Mode_Tendency
    df['Spend_Mode_Tendency'] = np.where(df['Online_Spend'] >= df['Offline_Spend'], 'Online', 'Offline')

    # Load phase: Since the task does not specify a destination, we return the transformed DataFrame
    return df

# Example of using the function:
# df_transformed = etl_pipeline('path_to_your_csv_file.csv')
```

### **1. Comportamento de compra do consumidor: análise da contribuição de vendas por localização geográfica**

**Análise comparativa da distribuição de vendas nos principais locais:**

Este código e o gráfico de visualização correspondente mostram a proporção de vendas em diferentes locais. A partir do gráfico de barras gerado, fica evidente que Chicago e Califórnia têm as maiores proporções de vendas, representando 34,81% e 30,88%, respectivamente. Em contraste, Washington DC tem a menor proporção de vendas, 5,48%. Essa análise pode ajudar a identificar quais locais estão apresentando bom desempenho em vendas e quais podem exigir mais esforços de marketing ou ajustes de estoque. O título do gráfico, “Proporção de vendas por local”, indica claramente que se trata de uma análise da distribuição de vendas por local.

```python
# Group by 'Location' and sum 'sub_sum', then calculate the sales proportion
sales_by_location = df_trans.groupby('Location', as_index=False)['sub_sum'].sum()
sales_by_location['Sales_Proportion'] = (sales_by_location['sub_sum'] /
                                         sales_by_location['sub_sum'].sum()) * 100

# Sort the data by 'Sales_Proportion' in descending order
sales_by_location = sales_by_location.sort_values('Sales_Proportion', ascending=False)

# Plotting the sales proportion for each location using seaborn
plt.figure(figsize=(8, 6))
ax = sns.barplot(data=sales_by_location, x='Location', y='Sales_Proportion', order=sales_by_location['Location'])

# Use the higher level bar_label method to add labels to the bars
ax.bar_label(ax.containers[0], fmt='%.2f%%', padding=3)  # Added padding for better label placement


plt.title('Sales Proportion by Location')
plt.ylabel('Sales Proportion (%)')
plt.xlabel('Location')

# Show the plot
plt.show()
```

```python
# Pivot the data to get sums for each gender within each location
df_pivoted = df_trans.pivot_table(index='Location', columns='Gender', values='sub_sum', aggfunc='sum', fill_value=0)

# Create a stacked bar plot
colors = ["pink", "#add8e6"]  # Colors for each gender
ax = df_pivoted.plot(kind='bar', stacked=True, color=colors, figsize=(8, 6))

# Add the percentage labels
for n, x in enumerate([*df_pivoted.index.values]):
    for (proportion, y_loc) in zip(df_pivoted.loc[x],
                                   df_pivoted.loc[x].cumsum()):
        ax.text(n, y_loc - (0.5 * proportion), f'{(proportion / df_pivoted.loc[x].sum() * 100):.0f}%',
                ha='center', va='center', color='black', fontweight='bold')

plt.title('Sales Performance by Location and Gender')
plt.ylabel('Sales Performance')
plt.xlabel('Location')

# Move the legend out of the plot
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

**Análise anual e trimestral da frequência de transações do cliente, extração e armazenamento da lista de clientes acima da frequência média de transações**

O objetivo desses dois blocos de código é analisar o comportamento das transações dos clientes e identificar os clientes cujas frequências de transações estão acima do nível médio.

Primeiro, é calculada a frequência média de transações por cliente durante todo o ano. Em seguida, esta frequência média anual é comparada com a frequência média de transações de cada trimestre para determinar quais clientes têm uma frequência de transações acima da média em cada trimestre e ao longo do ano.

Esses dados podem ser usados para análises posteriores, como identificar potenciais clientes fiéis ou avaliar a eficácia das estratégias de vendas. Ao salvar esses dados como um arquivo CSV, podemos processá-los facilmente ou compartilhá-los com outros membros da equipe.

```python
# Annual and Quarterly Customer Transaction Frequency Analysis

# First, we will calculate the average transaction frequency per customer for the entire year
annual_transaction_count_per_customer = df_trans.groupby('CustomerID').size()
annual_average_frequency = annual_transaction_count_per_customer.mean()

# Now we calculate the average transaction frequency per customer for each quarter
quarterly_transaction_count_per_customer = df_trans.groupby(['CustomerID', 'Quarter']).size().reset_index(name='TransactionCount')
quarterly_average_frequency_per_quarter = quarterly_transaction_count_per_customer.groupby('Quarter')['TransactionCount'].mean().to_dict()

# Creating lists for customers with above-average frequencies for each quarter and the year
above_average_customers_by_quarter = {quarter: [] for quarter in ['Q1', 'Q2', 'Q3', 'Q4', 'Year']}

# Loop over each customer and check their frequency against the average
for customer_id, transactions in annual_transaction_count_per_customer.items():
    if transactions > annual_average_frequency:
        above_average_customers_by_quarter['Year'].append(customer_id)

for (customer_id, quarter), data in quarterly_transaction_count_per_customer.groupby(['CustomerID', 'Quarter']):
    transactions = data['TransactionCount'].values[0]
    if transactions > quarterly_average_frequency_per_quarter[quarter]:
        above_average_customers_by_quarter[quarter].append(customer_id)

# Convert the dictionary of lists to a list of lists for the output
list_of_above_average_customers = [above_average_customers_by_quarter[quarter] for quarter in ['Q1', 'Q2', 'Q3', 'Q4', 'Year']]

# list_of_above_average_customers
```

```python
# Storage of Customer List Above Average Transaction Frequency

'''
Convert the dictionary of customers who are above the average frequency into a DataFrame
Each quarter will be a separate column, with the last column being the year
'''

# Create a DataFrame from the dictionary
df_above_average_customers = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in above_average_customers_by_quarter.items()]))

# Rename the columns for clarity
df_above_average_customers.rename(columns=lambda x: x if x != 'Year' else 'Annual', inplace=True)

df_above_average_customers.to_csv('above_average_customers.csv', index=False)

df_above_average_customers
```

**Análise trimestral e anual de receita de clientes de alta frequência**

O objetivo deste código é calcular a receita total de cada trimestre e de todo o ano, e compará-la com a receita aportada pelos clientes de alta frequência, aqueles cuja frequência de transação está acima da média. Ao fazer isso, podemos quantificar a contribuição desses clientes de alta frequência para a receita total e identificar quais períodos têm o impacto mais significativo na receita geral de vendas desses clientes valiosos.

```python
# Define a function to calculate total revenue for each quarter and the whole year, as well as revenue from customers with above-average frequency
def calculate_revenue(df, df_above_average_customers):
    # Initialize a dictionary to store total revenue and top customer revenue
    revenue_data = {
        'TOTAL REVENUE': [],
        'TOP CUSTOMER REVENUE': []
    }

    df_trans['Quarter'] = df_trans['Date'].dt.month.apply(month_to_quarter)

    # Ensure CustomerID in df_above_average_customers is of string type
    df_above_average_customers = df_above_average_customers.applymap(str)

    # Calculate total revenue for each quarter and the whole year
    for period in ['Q1', 'Q2', 'Q3', 'Q4', 'Year']:
        if period != 'Year':
            # For quarters, filter transactions for that quarter
            period_transactions = df_trans[df_trans['Quarter'] == period]
        else:
            # For the whole year, use all transactions
            period_transactions = df_trans

        # Calculate total revenue
        total_revenue = period_transactions['sub_sum'].sum()
        revenue_data['TOTAL REVENUE'].append(total_revenue)

        # Retrieve customer IDs with above-average frequency for the period
        if period != 'Year':
            top_customers = df_above_average_customers[period]
        else:
            top_customers = df_above_average_customers['Annual']

        # Calculate total revenue from top customers
        top_customer_revenue = period_transactions[period_transactions['CustomerID'].isin(top_customers)]['sub_sum'].sum()
        revenue_data['TOP CUSTOMER REVENUE'].append(top_customer_revenue)

    # Convert the dictionary to a DataFrame
    revenue_df = pd.DataFrame(revenue_data, index=['Q1', 'Q2', 'Q3', 'Q4', 'YEAR'])
    return revenue_df

# Use the function to calculate revenue
revenue_df = calculate_revenue(df_trans, df_above_average_customers)
revenue_df
```

A análise representada no gráfico ilustra uma descoberta significativa: os principais clientes representam uma maioria substancial da receita total em todos os trimestres e durante todo o ano. Especificamente, esses clientes contribuem com mais de 70% da receita total de vendas, com percentuais variando estreitamente de 71,04% a 72,84%. **Essa alta concentração de receitas de um seleto grupo de clientes ressalta sua importância para o negócio. Sugere uma forte dependência de uma base central de clientes e destaca o valor de cultivar esses relacionamentos. Esse padrão de gastos dos clientes também pode informar estratégias de marketing direcionadas e programas de fidelização de clientes para manter e potencialmente aumentar essas valiosas contribuições.**

```python

# Calculte the % of top customer revenue占total revenue of each period
revenue_df['PERCENTAGE'] = revenue_df['TOP CUSTOMER REVENUE'] / revenue_df['TOTAL REVENUE'] * 100

plt.figure(figsize=(8, 6))
plt.bar(revenue_df.index, revenue_df['PERCENTAGE'], color='skyblue')

# Add label
for i, percentage in enumerate(revenue_df['PERCENTAGE']):
    plt.text(i, percentage + 0.5, f'{percentage:.2f}%', ha='center', va='bottom')

plt.xlabel('Time Period')
plt.ylabel('Percentage (%)')
plt.title('Top Customer Revenue as a Percentage of Total Revenue')
plt.ylim(0, max(revenue_df['PERCENTAGE']) + 10)  # Give space
plt.show()
```

Sensibilidade do Consumidor: Os consumidores podem ser muito sensíveis aos preços. Mesmo com descontos, eles podem comprar apenas itens mais baratos ou tender a comprar em pequenas quantidades.

Estratégia de Desconto: Os descontos podem ser definidos em itens de menor preço, o que significa que mesmo que o consumidor compre mais por causa dos descontos, o valor médio de gasto permanece baixo.

Estrutura do produto: O comerciante pode oferecer uma variedade limitada de produtos com desconto ou esses produtos podem não ser atraentes o suficiente, resultando em um baixo volume de compras por parte dos consumidores.

Profundidade do desconto: Se a profundidade do desconto não for significativa, pode não ser suficiente para estimular os consumidores a aumentar o seu volume de compras, levando a um valor médio de gasto baixo.

Fidelização do cliente: A falta de fidelidade do cliente pode significar que os consumidores recorrem a outras marcas ou comerciantes em busca de descontos. Portanto, mesmo que um comerciante ofereça descontos, ele não pode garantir que atrairá consumidores suficientes.

### **9. Análise do Comportamento do Consumidor e Preferências de Clientes que Preferem Compras Online**

**Análise do impacto do gênero nas proporções de gastos em compras on-line e off-line**

Os resultados da análise indicam que, tanto para homens como para mulheres, a proporção de gastos offline (representados pela secção azul) é superior à dos gastos online (representados pela secção laranja).

```python
# Calculate the proportion of total online and offline spending per customer
df_trans['Total_Spend'] = df_trans['Offline_Spend'] + df_trans['Online_Spend']
df_trans['Offline_Spend_Prop'] = df_trans['Offline_Spend'] / df_trans['Total_Spend']
df_trans['Online_Spend_Prop'] = df_trans['Online_Spend'] / df_trans['Total_Spend']

# Calculate average proportion by gender group
gender_spend_props = df_trans.groupby('Gender')[['Offline_Spend_Prop', 'Online_Spend_Prop']].mean().reset_index()

plt.figure(figsize=(6, 4))
sns.barplot(data=gender_spend_props, x='Gender', y='Offline_Spend_Prop', color="blue", label='Offline Spend')
sns.barplot(data=gender_spend_props, x='Gender', y='Online_Spend_Prop', color="orange", label='Online Spend', bottom=gender_spend_props['Offline_Spend_Prop'])

plt.title('Average Online and Offline Spend Proportions by Gender')
plt.xlabel('Gender')
plt.ylabel('Proportion')
plt.legend(title='Spend Type')
plt.show()
```

**Identificando segmentos de clientes com preferência por compras on-line**

O objetivo deste código é segregar dos dados de consumo os clientes cujos gastos online excedem os gastos offline. Ao agregar os gastos online e offline de cada cliente, conseguimos identificar aqueles que têm preferência por compras online. Essas informações são particularmente úteis para compreender as preferências de compra dos clientes e podem ser usadas para orientar estratégias de marketing, melhorias no atendimento ao cliente e atividades promocionais personalizadas.

```python
# Aggregate online and offline spending for each customer
customer_total_spending = df_trans.groupby('CustomerID').agg({'Online_Spend': 'sum', 'Offline_Spend': 'sum'})

# Find customers who spend more online than offline
online_greater_customers = customer_total_spending[customer_total_spending['Online_Spend'] > customer_total_spending['Offline_Spend']]

online_greater_customers
```

**Análise das principais categorias de produtos para clientes que preferem compras online**

A intenção desta análise é obter insights sobre as categorias de produtos que os clientes que gastam mais online do que offline normalmente compram. Inicialmente, ele garante a consideração apenas dos IDs de cliente válidos. Posteriormente, os registros de compras desses clientes específicos são extraídos dos dados da transação. Após a confirmação de dados suficientes para análise, as cinco categorias de produtos mais frequentemente compradas por esses clientes são calculadas e apresentadas visualmente. Esta análise é inestimável para as empresas, pois pode ajudar a identificar potenciais oportunidades de vendas e permitir campanhas de marketing mais direcionadas ou decisões de inventário adaptadas a esses clientes.

```python
# Make sure online_greater_customers contains a valid customer ID
valid_customer_ids = online_greater_customers.index

# Extract the transaction records of these customers
valid_transactions = df_trans[df_trans['CustomerID'].isin(valid_customer_ids)]

# Check if there is enough data for analysis
if not valid_transactions.empty:
    # Calculate the top 5 product categories these customers purchase most frequently
    top_5_product_categories = valid_transactions['Product_Category'].value_counts().head(5)

    # Convert the results to a DataFrame for plotting
    df_top_5_product_categories = pd.DataFrame({'Product_Category': top_5_product_categories.index, 'Count': top_5_product_categories.values})

    plt.figure(figsize=(6, 4))
    sns.barplot(x='Count', y='Product_Category', data=df_top_5_product_categories)
    plt.title('Top 5 Product Categories for Online Greater Customers')
    plt.xlabel('Count')
    plt.ylabel('Product Category')
    plt.show()
else:
    print("There is insufficient data for analysis.")
```

Os resultados da análise mostram que estes dois tipos de comportamento do consumidor, clientes predominantes que gastam online e clientes em geral, não diferem significativamente.
